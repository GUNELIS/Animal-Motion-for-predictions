```python
def find_closest_row(meteor, event_a):
    best_date = None
    shortest_dif = pd.Timedelta(days=100000)
    for date in meteor['date']:
        dif = abs(event_a - date)
        if dif <= shortest_dif:
            shortest_dif = dif
            best_date = date
    
    closest_row = meteor[meteor['date'] == best_date]  # Get the row with the closest date
    return closest_row

def add_bar_pressure(baro2015,animals):
    for index, row in animals.iterrows():
        closest_row = find_closest_row(baro2015,row['date'])
        animals.at[index, 'Barometric Pressure'] = closest_row['Barometric Pressure (mb)'].iloc[0]  # Add the entry from closest_row to animal_df
    return animals

def add_Wind_speed(windSpeed,animals):
    for index, row in animals.iterrows():
        closest_row = find_closest_row(windSpeed,row['date'])
        animals.at[index, 'Wind Speed (kn)'] = closest_row['Wind Speed (kn)'].iloc[0]  # Add the entry from closest_row to animal_df
    return animals

def add_Air_temp(air_temp,animals):
    for index, row in animals.iterrows():
        closest_row = find_closest_row(air_temp,row['date'])
        animals.at[index, 'Air Temp (°F)'] = closest_row['Air Temp (°F)'].iloc[0]  # Add the entry from closest_row to animal_df
    return animals

def add_Wind_Gust(wind_gust,animals):
    for index, row in animals.iterrows():
        closest_row = find_closest_row(wind_gust,row['date'])
        animals.at[index, 'Wind Gust (kn)'] = closest_row['Wind Gust (kn)'].iloc[0]  # Add the entry from closest_row to animal_df
    return animals

def add_humidity(humidity,animals):
    for index, row in animals.iterrows():
        closest_row = find_closest_row(humidity,row['date'])
        animals.at[index, 'Humidity (%)'] = closest_row['Humidity (%)'].iloc[0]  # Add the entry from closest_row to animal_df
    return animals
```
