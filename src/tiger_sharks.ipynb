```python
import utilities as util
import preprocessor as prep
import pandas as pd
import folium
```

```python
full_hurricane_1992 = prep.process_hurricane_txt('C:/Users/Ben/Documents/GitHub/Animal-Motion-for-predictions/data/hurricanes_1992.txt')
tiger_sharks = prep.preprocess_tiger_sharks('C:/Users/Ben/Documents/GitHub/Animal-Motion-for-predictions/data/whaleSharks/MAjemian_TigerShark_SPOT_locs_2010-2018.csv')
g_p = util.get_cross_points(full_hurricane_1992,tiger_sharks,350)
```

```python
map = folium.Map(location = [20,-85],zoom_start=3)
grouped_event_name = g_p.groupby('event_name')
grouped_hurri_name = full_hurricane_1992.groupby('name')
i = -1
for name, tiger_sharks in grouped_event_name:
    i+=1
    map = util.mark_map(map,tiger_sharks,'fish',util.color_list[i])
    map = util.mark_map(map,grouped_hurri_name.get_group(name),'tornado',util.color_list[i])
```
